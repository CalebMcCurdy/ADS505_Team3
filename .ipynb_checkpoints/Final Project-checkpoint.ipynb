**Numerical Variables vs Categorical Variables**

```python
# differentiating which variables are numeric vs categorical
num_vars = ['age', 'balance','day','duration','campaign','pdays','previous']
cat_vars = list(bank.drop(columns = num_vars, axis = 1).columns)

# updating the object variables to categorical
bank[cat_vars] = bank[cat_vars].astype('category') 
```

**Statistics of those who subscribed a term deposit** 

```python
# creating a subset of data for those who subscribed
bank_yes = bank[bank['y'] == 'yes']

# getting the statistics on the numerical values
bank_yes.describe().T
```

**Correlations between the numeric variables**

```python
# heatmap of the numerical values
sns.heatmap(bank[num_vars].corr())
```

Not many numerical variables are correlated with each other. But, it looks like pdays and previous are highly correlated.

**Scatterplots of just the numeric variables**

**How many clients did or did not subscribe to a term deposit based on some of the categorical variables that seemed interesting**

```python
#setting up plots
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 13))

# add suptitle
fig.suptitle('Clients that subscribed to a term deposit')

# plotting by job
ax0 = pd.crosstab(bank['job'],bank['y']).plot.bar(stacked=True, ax=axes[0][0] )
ax0.set_title('Responses based on Job')

# plotting by marital status
ax1 = pd.crosstab(bank['marital'],bank['y']).plot.bar(stacked=True, ax=axes[0][1])
ax1.set_title('Responses based on Marital Status')

# plotting by education
ax2 = pd.crosstab(bank['education'],bank['y']).plot.bar(stacked=True, ax=axes[1][0])
ax2.set_title('Responses based on Education')

# plotting by default status
ax3 = pd.crosstab(bank['default'],bank['y']).plot.bar(stacked=True, ax=axes[1][1])
ax3.set_title('Responses based on posessesion of default loans')

# plotting by housing loan
ax4 = pd.crosstab(bank['housing'],bank['y']).plot.bar(stacked=True, ax=axes[2][0] )
ax4.set_title('Responses based on possession of housing loans')

# plotting by personal loan
ax5 = pd.crosstab(bank['loan'],bank['y']).plot.bar(stacked=True, ax=axes[2][1])
ax5.set_title('Responses based on posesssion of personal loans')

# plotting by contact type
ax6 = pd.crosstab(bank['contact'],bank['y']).plot.bar(stacked=True, ax=axes[3][0])
ax6.set_title('Responses based on communication type')

# plotting by last contact month
ax7 =pd.crosstab(bank['poutcome'],bank['y']).plot.bar(stacked=True, ax=axes[3][1])
ax7.set_title('Responses based on last campaign outcome')

fig.set_figheight(15)
fig.set_figwidth(15)
fig.subplots_adjust(top = .9, hspace=.90)

plt.show()
```

**Boxplots for numeric variables vs y**

```python
#setting up plots
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 13))
num_vars = ['age', 'balance','day','duration','campaign','pdays','previous']
# add suptitle
fig.suptitle('Boxplots of numeric variables vs y')

# plotting by job
ax0 = sns.boxplot(x = 'age', y = 'y', data = bank, ax = axes[0][0])
ax0.set_title('Responses based on Age')

ax1 = sns.boxplot(x = 'balance', y = 'y', data = bank, ax = axes[0][1])
ax1.set_title('Responses based on Average yearly balance')

ax2 = sns.boxplot(x = 'day', y = 'y', data = bank, ax = axes[1][0])
ax2.set_title('Responses based on last contact day of the month')

ax3 = sns.boxplot(x = 'duration', y = 'y', data = bank, ax = axes[1][1])
ax3.set_title('Responses based on last contact duration, in seconds')

ax4 = sns.boxplot(x = 'campaign', y = 'y', data = bank, ax = axes[2][0])
ax4.set_title('Responses based on number of contacts performed \n during this campaign for this client')

ax5 = sns.boxplot(x = 'pdays', y = 'y', data = bank, ax = axes[2][1])
ax5.set_title('Responses based on number of days that passed by\n after the client was last contacted from a previous campaign')

ax7 = sns.boxplot(x = 'previous', y = 'y', data = bank, ax = axes[3][0])
ax7.set_title('Responses based on number of contacts performed before \n this campaign and for this client')

fig.set_figheight(15)
fig.set_figwidth(15)
fig.subplots_adjust(top = .9, hspace=.90)
```
